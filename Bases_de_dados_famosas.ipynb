# **Base de dados *Íris***

Esses conjuntos de dados consistem em 3 tipos diferentes de comprimento de pétala e sépala de íris (Setosa, Versicolor e Virginica), armazenados em um numpy.ndarray de ordem 150x4, sendo as linhas as amostras e as colunas: Comprimento da Sépala, Largura da Sépala, Comprimento da Pétala e Largura da Pétala.

O gráfico abaixo usa os dois primeiros recursos. Veja [aqui](https://en.wikipedia.org/wiki/Iris_flower_data_set) mais informações sobre este conjunto de dados.

```python
# Código fonte: Gaël Varoquaux
# Modificado para documentação por Jaques Grobler
# Adaptado para publicação no GitHub por Rafael Barros
# Licença: BSD 3 clause

import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA

# iImportação dos dados
iris = datasets.load_iris()
X = iris.data[:, :2]  # vamos adotar apenas as duas primeiras features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plotagem dos pontos de treinamento
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Comprimento da sépala")
plt.ylabel("Largura da sépala")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

#Para ter uma melhor compreensão da relação entre as dimensões,
#plota-se as três primeiras dimensões dos componentes de análise
#principais (Principal Component Analysis, PCA)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduzido = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduzido[:, 0],
    X_reduzido[:, 1],
    X_reduzido[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title(" Três primeiras direções PCA")
ax.set_xlabel("1º autovetor")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2º autovetor")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3º autovetor")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
